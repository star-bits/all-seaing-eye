```python
import xmltodict
import json
import os
from glob import glob 
```

```python
d = {}

d['info'] = {
    'year': 2022, 
    'version': '1.3', 
    'description': '해상 객체 이미지', 
    'contributor': '한국해양과학기술원부설선박해양플랜트연구소', 
    'url': 'https://aihub.or.kr/aidata/34155', 
    'date_created': '2022-01-10'
}

d['licenses'] = [{
    'id': 0, 
    'name': 'CC BY 4.0', 
    'url': 'https://creativecommons.org/licenses/by/4.0/deed.ast'
}]

d['images'] = []
d['categories'] = [
    {
        'id': 2,
        'name': '선박',
        'supercategory': '선박'
    },
    {
        'id': 3,
        'name': '기타부유물',
        'supercategory': '기타부유물'
    },
    {
        'id': 1,
        'name': '어망부표',
        'supercategory': '어망부표'
    },
]
d['annotations'] = []
```

```python
l = []

img_lbl_dir = 'train/labels'

for date_dir in glob(os.path.join(img_lbl_dir, '*')):

    for vid_dir in glob(os.path.join(date_dir, '*')):

        for file_path in glob(os.path.join(vid_dir, '*.xml')):
            l.append(file_path)
            
print(len(l))
```

```python
# l
```

```python
from tqdm import tqdm

category_ids = []
supercategories = []
categories = []

for itr, file_path in tqdm(enumerate(l)):
    xml_path = file_path
    with open(xml_path, 'r', encoding='utf-8') as f:
        xml_info = f.read()
    xml_info = xmltodict.parse(xml_info)
    
    img_info = {}
    file_name = file_path.replace('labels', 'images')
    file_name = file_name.replace('.xml', '.jpg')
    img_info['file_name'] = file_name    
    img_info['width'] = int(xml_info['annotation']['size']['width'])
    img_info['height'] = int(xml_info['annotation']['size']['height'])
    img_info['license'] = 0

    try: 
        xml_info['annotation']['object']
    except KeyError:
        img_id = int(xml_info['annotation']['filename'][-22:-14] + xml_info['annotation']['filename'][-13:-9] + xml_info['annotation']['filename'][-8:-4])
        print(img_id)
        img_info['id'] = img_id
        d['images'].append(img_info)
        continue
    else:   
        if str(type(xml_info['annotation']['object']))=="<class 'dict'>":
            ann_info = {}
            img_id = int(xml_info['annotation']['object']['image_id'])
            ann_info['image_id'] = img_id
            ann_info['id'] = int(xml_info['annotation']['object']['id'])
            ann_info['category_id'] = int(xml_info['annotation']['object']['category_id'])
            ann_info['area'] = float(xml_info['annotation']['object']['area'])
            ann_info['bbox'] = list(map(int, xml_info['annotation']['object']['bbox'][1:-1].split(',')))
            ann_info['iscrowd'] = 0
            d['annotations'].append(ann_info)
            
            if ann_info['category_id'] not in category_ids:
                category_ids.append(ann_info['category_id'])
                supercategories.append(xml_info['annotation']['object']['property']['supercategory'])
                categories.append(xml_info['annotation']['object']['property']['category'])

        else:
            for i in range(len(xml_info['annotation']['object'])):
                ann_info = {}
                img_id = int(xml_info['annotation']['object'][i]['image_id'])
                ann_info['image_id'] = img_id
                ann_info['id'] = int(xml_info['annotation']['object'][i]['id'])
                ann_info['category_id'] = int(xml_info['annotation']['object'][i]['category_id'])
                ann_info['area'] = float(xml_info['annotation']['object'][i]['area'])
                ann_info['bbox'] = list(map(int, xml_info['annotation']['object'][i]['bbox'][1:-1].split(',')))
                ann_info['iscrowd'] = 0
                d['annotations'].append(ann_info)
                
                if ann_info['category_id'] not in category_ids:
                    category_ids.append(ann_info['category_id'])
                    supercategories.append(xml_info['annotation']['object'][i]['property']['supercategory'])
                    categories.append(xml_info['annotation']['object'][i]['property']['category'])

    img_info['id'] = img_id
    d['images'].append(img_info)
```
